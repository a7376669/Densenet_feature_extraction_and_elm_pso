```python
from sko.GA import GA
from sko.PSO import PSO
#ELM 算法（regreesion）
def elm_function(x):
    # -----------------training---------------------#
    # input weight
    n_input_weights=100
    input_weights = np.array(x[0:n_input_weights])
    # bias
    biases = np.array(x[n_input_weights:])
    # menghitung matriks keluaran hidden layer pada data training
    n_hidden_node=100
    #讀資料進Data.dt_training
    hidden_layer_training = Hidden_layer(input_weights, biases, n_hidden_node, Data.dt_training)
    activation_training = Activation(hidden_layer_training)
    pseudo_training = Pseudoinverse(activation_training)
    #Data.dt_target_training目標標籤
    output_training = Output_weight(pseudo_training, Data.dt_target_training)

    # -----------------testing--------------------#
    hidden_layer_testing = Hidden_layer(input_weights, biases, n_hidden_node, Data.dt_testing)
    activation_testing = Activation(hidden_layer_testing)
    # prediction = Target_output(hidden_layer_testing, output_training)
    prediction = Target_output(activation_testing, output_training)
    # akurasi
    accuracy = Evaluate(Data.dt_target_testing, prediction, t='xyz')

    return accuracy

def Hidden_layer(input_weights, biases, hidden_nodes, data_input):
    input_weight = input_weights.reshape(hidden_nodes, feature_dim)
    bias = biases
    transpose_input_weight = np.transpose(input_weight)
    hidden_layer = np.add(np.matmul(data_input, transpose_input_weight), bias)
    return hidden_layer

# 活化函數
def Activation(hidden_layer):
    # activation = 1 / (1 + np.exp(-hidden_layer))
    activation = .5 * (1 + np.tanh(.5 * hidden_layer))
    return activation


def Pseudoinverse(hidden_layer):
    h_pseudo_inverse = scipy.linalg.pinv(hidden_layer, cond=None, rcond=None, return_rank=False, check_finite=True)
    return h_pseudo_inverse

# 輸出權重
def Output_weight(pseudo_inverse, target):
    beta = np.matmul(pseudo_inverse, target)
    return beta


def Target_output(testing_hidden_layer, output_weight):
    prediction = np.matmul(testing_hidden_layer, output_weight)
    return prediction


def Evaluate(actual, prediction, t='xyz'):
    accuracy = np.square(actual - prediction)
    return np.mean(np.sum(accuracy, axis=1))



# 獲得ELM的訓練參數，beta，w，b
def elm_parameter(x):
    input_weights = np.array(x[0:n_input_weights])
    biases = np.array(x[n_input_weights:])

    hidden_layer_training = Hidden_layer(input_weights, biases, n_hidden_node, Data.dt_training)
    activation_training = Activation(hidden_layer_training)
    pseudo_training = Pseudoinverse(activation_training)

    w = np.transpose(input_weights.reshape(n_hidden_node, feature_dim))
    b = np.array(biases)
    beta = Output_weight(pseudo_training, Data.dt_target_training)
    return beta, w, b

# 輸入數據，得到預測輸出
def elm_output(data, w, b, beta):
    temp = np.add(np.matmul(data, w), b)
    activation_temp = Activation(temp)
    output = np.dot(activation_temp, beta)
    out_xyz = np.multiply(output, [5000, 5000, 3000])
    return out_xyz
```

```python
#參數定義 
n_dim=100
population=40
max_iter=150
#pso優化elm
pso = PSO(func=elm_function, n_dim=n_dim, pop=population, max_iter=max_iter, w=0.6, c1=1.5,
          lb=[-1] * n_dim, ub=[1] * n_dim, c2=1.5)
```
